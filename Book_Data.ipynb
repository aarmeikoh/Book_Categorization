```python
import requests
import random
import pandas as pd
import re
from bs4 import BeautifulSoup as BS
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
import matplotlib.pyplot as plt
```

```python
books = {}
n = 0
nums = []
while len(books) < 2000:
    num = random.randint(0,69999)
   
    if num in nums:
        continue
    else:
        nums.append(num)
    # num = 35304
    url = f'https://www.gutenberg.org/files/{num}/{num}-h/{num}-h.htm'
    result = requests.get(url, verify= False)
    if result.status_code == 200:
        try:
            soup = BS(result.text)
            title = soup.find('title').text
            book = {}
            body = soup.get_text().strip()
            text = body[1000: len(body)-19300]
            words = text.split()
            author = re.findall('(?<=by).*', title)[0].strip()
            print(num, len(words))
            if author in ['Various.', 'Various', '', 'AUTHOR.', 'Anonymous', 'Anonymous.', 'Various Authors.'] or len(words) < 3500:
                continue
            book['name'] = re.findall('(?<=of).*(?=,)', title)[0].strip()
            book['author'] = author
            body = soup.get_text().strip()
            text = body[1000: len(body)-19300]
            words = text.split()
            start = random.sample(range(len(words)-600),k= 10) # -500 added
            book['url'] = url
            book['length'] = len(words)

            for i, a in enumerate(start):
                book[f'sample{i}'] = ' '.join(words[a:a+550])
        except:
            continue
    else:
        continue
    books[n] = book
    n += 1
```
